```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data_path = "../data/"
frq = pd.read_csv(data_path+'frequency.csv')
per = pd.read_csv(data_path+'perimeters.csv')
per.head()
frq.head()

```

```python
list(frq)
```

```python
# remove unneeded columns
columns = ['ASMNT_TYPE', 'PRE_ID', 'POST_ID', 'ND_T', 'IG_T', 'LOW_T', 'MOD_T', 'HIGH_T', 'WRS_PATH', 'WRS_ROW','R_ACRES', 'ADMIN','MTBS_ZONE', 'GACC', 'HUC4_NAME', 'Version',  'RevCode', 'RelDate', 'HUC4_CODE']
frq = frq.drop(columns, 1)
frq.head()
```

```python
frq.dtypes
```

```python
# create new column for datetime
frq.insert(5, 'datetime_str', str)
```

```python

# convert to int
conversion_colmuns = ['FIRE_YEAR', 'FIRE_MON', 'FIRE_DAY']
for item in conversion_colmuns:
    frq[item] = frq[item].astype(np.int64)
        
# add leading zeros
frq['FIRE_MON'] = frq.FIRE_MON.map('{:02}'.format)
frq['FIRE_DAY'] = frq.FIRE_DAY.map('{:02}'.format)

# convert back to strings
for item in conversion_colmuns:
    frq[item] = frq[item].astype(np.str)

frq['datetime_str'] = frq.FIRE_YEAR + '-' + frq.FIRE_MON + '-' + frq.FIRE_DAY + ' 00:00:00' 

frq.dtypes
frq.head()
```

```python
# convert to datetime
frq.datetime_str = pd.to_datetime(frq.datetime_str)
frq.head()
```

```python
# join and print 

wildfires = pd.merge(frq, per, on ="Fire_ID")
# columns = ['Fire_Name', 'Year']
# frq = frq.drop(columns, 1)
frq.head()
```

```python
wildfires.head()
wildfires.dtypes
```

```python
# sort to the last 25 years
wf = wildfires[wildfires['Year'] > 1996]
wf = wf[wf['Fire_Type'] == 'WF']

wf.head()
```

```python
wf.to_csv(data_path+'wildfires.csv')
```
